# Data description

The dataset used in this project is the Aficionado Coffee Roasters sales dataset, which contains transactional-level data collected from café store operations. The dataset captures detailed information related to customer purchases, product characteristics, pricing, and store locations. It consists of 11 columns, each representing a specific aspect of a transaction. This structured data enables comprehensive analysis of sales behavior, product performance, and time-based demand patterns across multiple beverage categories.

A detailed description of each column is provided below:

# Column Description

1. transaction_id :
A unique identifier assigned to each transaction. This column ensures that every sales record can be uniquely distinguished and is useful for tracking and auditing individual purchases.

2. year :
Represents the calendar year in which the transaction occurred. This column helps in analyzing long-term sales trends and performing year-wise comparisons.

3. transaction_time :
Indicates the exact time at which the transaction took place. Initially stored as an object (string) data type, this column was later converted into a datetime format to enable time-based analysis such as identifying peak sales hours and daily demand patterns.

4. transaction_qty :
Refers to the quantity of items sold in a single transaction. This variable is essential for understanding sales volume, demand intensity, and customer purchasing behavior.

5. store_id :
A unique identifier for each store or café outlet. This column allows for store-level analysis and comparison of performance across different locations.

6. store_location :
Specifies the geographical location of the store where the transaction occurred. It supports spatial analysis and helps identify location-based differences in sales performance.

7. product_id :
A unique identifier assigned to each product sold by the café. This column enables product-level tracking and analysis.

8. unit_price :
Represents the selling price per unit of the product at the time of the transaction. This variable is crucial for revenue calculation and pricing analysis.

9. product_category :
Indicates the broader category to which a product belongs, such as coffee, tea, or drinking chocolate. This column allows analysis at a category level to understand the contribution of different beverage types to overall sales.

10. product_type :
Provides a more detailed classification within each product category. For example, different types of coffee or tea can be distinguished using this column, enabling granular product-level insights.

11. product_details :
Contains descriptive information about the product, such as flavor, blend, or specific variant. This column adds contextual detail and can be useful for advanced analysis related to customer preferences.

# problem statement

One of the major challenges identified in the dataset was the representation of the transaction time variable. The transaction time was originally stored as an object (string) data type, which restricted its direct use in time-based analysis and forecasting. Without converting this variable into a proper datetime format, it was not possible to extract meaningful temporal insights or perform accurate time-series modeling.

To address this issue, the transaction time column was converted into a datetime object using appropriate data preprocessing techniques. This conversion enabled the extraction of additional time-based features, such as the transaction date and the hour of transaction. These derived features played a crucial role in understanding daily sales trends, identifying peak business hours, and analyzing customer purchasing behavior across different time periods.

The creation of separate date and hour columns significantly enhanced the analytical depth of the project. It allowed for the aggregation of sales data at different temporal levels and supported more accurate demand forecasting by incorporating time-dependent patterns. This preprocessing step was essential for enabling reliable exploratory data analysis, feature engineering, and predictive modeling within the café sales forecasting framework.

# Methodology

The methodology of this project follows a systematic data science workflow designed to analyze historical café sales data and develop a reliable forecasting framework. The process consists of data understanding, preprocessing, exploratory analysis, feature engineering, model development, and performance evaluation. All analysis and modeling were carried out using Python in Google Colab.

1. Data Understanding : The project began with an initial examination of the Aficionado Coffee Roasters dataset to understand its structure, data types, and key variables. The dataset contains transaction-level information related to sales quantity, product characteristics, pricing, store location, and transaction timing.

2. Data Preprocessing : Data preprocessing was a crucial step to ensure data quality and usability. One of the main challenges was that the transaction_time column was stored as an object data type. This column was converted into a datetime format to enable time-based analysis. From the converted datetime variable, additional features such as transaction date and transaction hour were extracted.

    Other preprocessing steps included:

    * Checking and handling missing or duplicate records

    * Ensuring numerical columns such as quantity and unit price were correctly typed

    * Creating derived variables such as total sales revenue where required

3. Exploratory Data Analysis (EDA) : Exploratory Data Analysis was conducted to uncover patterns and insights within the data. Visualizations and summary statistics were used to analyze:

    * Revenue by product category and store location.
    * Distribution of transaction quantity.
    * top 10 product by revenue.
    * hourly and daily demand patterns.

    EDA helped in identifying seasonality, peak sales hours, and the relative contribution of different beverage categories to overall revenue.

4. Feature Engineering : Based on insights from EDA, relevant features were selected and engineered to improve model performance. Time-based features such as year, date, and hour were used to capture temporal effects. Categorical variables like product category, product type, and store location were encoded appropriately for modeling. Feature engineering ensured that the models could effectively learn patterns from both temporal and product-level data.

5. Model Development : Predictive models were developed to forecast sales using historical data. A time-based train–test split was employed instead of random splitting to prevent data leakage and to replicate real-world forecasting scenarios. Multiple regression and machine learning models were evaluated to predict revenue.(In this model we train two madel linear regression or random forrest regression)

6. Model Evaluation : Model performance was assessed using standard evaluation metrics such as Mean Squared Error (MSE) and R-squared (R²). These metrics were used to compare model predictions against actual sales values and to select the best-performing model. Rolling or forward validation techniques were also applied to further test the robustness of the models over time.

7. Tools and Technologies : The entire project was implemented using Python in Google Colab. Key libraries used include:

    * Pandas and NumPy for data manipulation

    * Matplotlib for data visualization

    * Scikit-learn for model building and evaluation
